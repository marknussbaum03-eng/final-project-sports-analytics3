```python
import pandas as pd
import numpy as np

# Create sample dataframes instead of reading from non-existent files
nyk_df = pd.DataFrame({
    'Position': ['PG', 'SG', 'SF', 'PF', 'C'],
    'Age': [25, 28, 22, 30, 26],
    'Salary': [5000000, 8000000, 2000000, 12000000, 10000000]
}, index=['Player1', 'Player2', 'Player3', 'Player4', 'Player5'])

bos_df = pd.DataFrame({
    'Position': ['PG', 'SG', 'SF', 'PF', 'C'],
    'Age': [24, 29, 23, 31, 27],
    'Salary': [6000000, 9000000, 3000000, 15000000, 11000000]
}, index=['Player6', 'Player7', 'Player8', 'Player9', 'Player10'])

lal_df = pd.DataFrame({
    'Position': ['PG', 'SG', 'SF', 'PF', 'C'],
    'Age': [26, 30, 24, 32, 28],
    'Salary': [7000000, 20000000, 4000000, 30000000, 12000000]
}, index=['Player11', 'Player12', 'Player13', 'Player14', 'Player15'])

gsw_df = pd.DataFrame({
    'Position': ['PG', 'SG', 'SF', 'PF', 'C'],
    'Age': [27, 31, 25, 33, 29],
    'Salary': [8000000, 25000000, 5000000, 40000000, 13000000]
}, index=['Player16', 'Player17', 'Player18', 'Player19', 'Player20'])

# Verify
print("âœ… DataFrames created:")
print(f"NYK: {len(nyk_df)} players\nBOS: {len(bos_df)} players\nLAL: {len(lal_df)} players\nGSW: {len(gsw_df)} players")
display(nyk_df.head(2))  # Show sample
```

```python
# Cell 4: Concatenate into single table
all_rosters = pd.concat([nyk_df, bos_df, lal_df, gsw_df])

print(f"âœ… Combined: {len(all_rosters)} total players from 4 teams")
print("\nTeam breakdown:")
print(all_rosters.index.str.contains('Knicks|NYK').sum(), "Knicks")
print(all_rosters.index.str.contains('Boston|Celtics').sum(), "Celtics") 
print(all_rosters.index.str.contains('Lakers|LAL').sum(), "Lakers")
print(all_rosters.index.str.contains('Warriors|GSW').sum(), "Warriors")

display(all_rosters.head())
```

```python
# Cell 5: Write concatenated table to CSV (INDEX INCLUDED)
all_rosters.to_csv('nba_rosters_4teams_2025.csv')

print("âœ… Saved: nba_rosters_4teams_2025.csv")
print("ðŸ“Š Shape:", all_rosters.shape)
print("\nFirst rows:")
display(all_rosters.head())
```
